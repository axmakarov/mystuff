```python
import pandas as pd
from os import listdir
from os.path import isfile, join
mypath = u'C:\\Users\\aleksey.makarov\\Desktop\\2016-03\\Дерево стратегий\\Логи консольного приложения\\'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
channelsDFs = []
checkersDFs = []
deletedChannelsDFs = []
for f in onlyfiles:
    print f
    with open(mypath+f,'r+') as f1:
        df = pd.read_csv(f1,encoding='cp1251',sep='\t')
        if 'checkersInfo' in f:
            checkersDFs.append(df)
        if 'channelsForUseInfo' in f:
            channelsDFs.append(df)
        if 'deletedChannelInfo' in f:
            deletedChannelsDFs.append(df)
print len(channelsDFs)
print len(checkersDFs)
print len(deletedChannelsDFs)
```

```python
df_channels = pd.concat(channelsDFs)
df_checkers = pd.concat(checkersDFs)
df_deleted_channels = pd.concat(deletedChannelsDFs)
```

```python
df_channels.to_excel(u'C:\\Users\\aleksey.makarov\\Desktop\\2016-03\\Дерево стратегий\\2016-03-11_stree_channels.xlsx')
df_checkers.to_excel(u'C:\\Users\\aleksey.makarov\\Desktop\\2016-03\\Дерево стратегий\\2016-03-11_stree_checkers.xlsx')
df_deleted_channels.to_excel(u'C:\\Users\\aleksey.makarov\\Desktop\\2016-03\\Дерево стратегий\\2016-03-11_stree_deleted_channels.xlsx')
```
