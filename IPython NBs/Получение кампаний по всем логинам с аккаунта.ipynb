```python
import json
import requests
import pandas as pd
import re
```

```python
url = 'https://api.direct.yandex.ru/live/v4/json/'
payload = {
   "method": "GetClientsList",
    "token": "cc8fb959428d401a98ab953b75693f2e"
   }
jpayload = json.dumps(payload,ensure_ascii=False).encode('utf-8')
data = json.loads(requests.post(url,jpayload).text)
```

```python
goodLogins = [login for login in data['data'] if 'ClientCurrencies' in login.keys() and login['ClientCurrencies'] == [u'RUB']]
```

```python
loginsBatches = [goodLogins[i:i+100] for i in xrange(0,len(goodLogins),100)]
```

```python
i = 1
cnt = len (loginsBatches)
campaigns = []
for loginsBatch in loginsBatches:
    print '{0} of {1}'.format(i,cnt)
    i += 1
    payload = {
        "method": "GetCampaignsList",
        "token": "cc8fb959428d401a98ab953b75693f2e",
        "param": [login['Login'] for login in loginsBatch]
    }
    jpayload = json.dumps(payload,ensure_ascii=False).encode('utf-8')
    cmpgns = json.loads(requests.post(url,jpayload).text)
    campaigns += cmpgns['data']
```

```python
campaignsBatches = [campaigns[i:i+100] for i in xrange(0,len(campaigns),100)]
```

```python
i = 1
cnt = len(campaignsBatches)
fullCampaigns = []
for campaignsBatch in campaignsBatches:
    print '{0} of {1}'.format(i,cnt)
    i += 1
    payload = {
        "method": "GetCampaignsParams",
        "token": "cc8fb959428d401a98ab953b75693f2e",
        "param": {
            "CampaignIDS": [campaign['CampaignID'] for campaign in campaignsBatch],
            "Currency": "RUB"
        }
    }
    jpayload = json.dumps(payload,ensure_ascii=False).encode('utf-8')
    cmpgns = json.loads(requests.post(url,jpayload).text)
    fullCampaigns += cmpgns['data']
```

```python
pd.DataFrame(fullCampaigns).to_csv('D:\\allCampaigns.csv',encoding='utf-8',sep='\t',decimal=',')
```
