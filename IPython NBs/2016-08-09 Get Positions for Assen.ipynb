```python
import json,requests
```

```python
def createPayload (query,lr):
    payload = {
     "jsonrpc": "2.0",
     "method": "getYandexPosition",
     "params": {
             "data": {
                 "lr": str(lr),
                 "query": query,
                 "zone": "ru"
             },
             "token": "UH7ForyW2o2QGIruyktzXbiivQ89EsEK197Ioccc"
     },
     "id":123
}
    return json.dumps(payload)
```

```python
url = 'http://seozoo.ru/api'
data = requests.post(url,createPayload(u'купить авто',213))
```

```python
json.loads(data.text)['result']['results']
```

```python
def processResponse(data,query):
    results = json.loads(data.text)['result']['results']
    for result in results:
        result['query'] = query
    return results
```

```python
lr = 213
query = u'купить автомобиль'
data = requests.post(url,createPayload(query,lr))
results = processResponse(data,query)
```

```python
import pandas as pd
```
