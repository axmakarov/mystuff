```python
for item in olapData:
    dimensions = item['dimensions']
    item['Phrase'] = dimensions[0]
    item['Query'] = dimensions[1]
    metrics = item['value'][0]
    item['Visits'] = metrics['visits']
    item['CPAVisits'] = metrics['cpaVisits']
    item['AvgVisitDuration'] = metrics['avgVisitDuration']
    item['BounceRate'] = metrics['rejectPercent']
    item['PageDepth'] = metrics['pageDepth']
    item['SumVisitDuration'] = item['AvgVisitDuration']*item['Visits']
    item['Bounces'] = item['BounceRate']*item['Visits']/100
    item['Pageviews'] = metrics['pageDepth']*item['Visits']
```

```python
morph = pymorphy2.MorphAnalyzer()
for item in cleanOlapData:
    item['CleanedPhrase'] = SortWords(ClearQuery(CleanStopWords(item['Phrase'])))
    item['CleanedQuery'] = SortWords(ClearQuery(CleanStopWords(item['Query'])))
    item['NormalizedPhrase'] = ' '.join([morph.parse(word)[0].normal_form for word in item['CleanedPhrase'].split(' ')])
    item['NormalizedQuery'] = ' '.join([morph.parse(word)[0].normal_form for word in item['CleanedQuery'].split(' ')])
    item['IsPhraseEqualsQuery'] = (item['NormalizedPhrase'] == item['NormalizedQuery'])
```

```python
from collections import defaultdict
WordsVisitsDict = defaultdict(int)
WordsCPAVisitsDict = defaultdict(int)
WordsSumVisitDurationDict = defaultdict(int)
WordsBouncesDict = defaultdict(int)
WordsPageviewsDict = defaultdict(int)
for item in cleanOlapData:
    for word in item['WordsArray']:
        WordsVisitsDict[word] += item['Visits']
        WordsCPAVisitsDict[word] += item['CPAVisits']
        WordsSumVisitDurationDict[word] += item['SumVisitDuration']
        WordsBouncesDict[word] += item['Bounces']
        WordsPageviewsDict[word] += item['Pageviews']
```

```python
wordsData = []
for k,v in WordsVisitsDict.iteritems():
    d = {}
    d['Word'] = k
    d['Visits'] = v
    d['CPAVisits'] = WordsCPAVisitsDict[k]
    d['SumVisitDuration'] = WordsSumVisitDurationDict[k]
    d['Bounces'] = WordsBouncesDict[k]
    d['Pageviews'] = WordsPageviewsDict[k]
    wordsData.append(d)
```
