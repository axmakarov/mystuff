```python
from os import listdir
from os.path import isfile, join
mypath = u'\\\CHERNIKOVA-IDEV\\Share\\CarmaNewDomain\\ReplaceSpace\\'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
all_df = pd.DataFrame()
for f in onlyfiles:
    df = pd.read_csv(mypath+f,sep=';',encoding='cp1251')
    df = df[~df[u'Domain'].isin([u'url'])]
    df['file'] = f
    all_df = pd.concat([df,all_df])
```

```python
all_df['Price1'] = all_df['Price1'].astype(float)
all_df['Price2'] = all_df['Price2'].astype(float)
```

```python
keywords_pivot = all_df[['Keyword','Price1','Price2','file']].groupby(['Keyword','file'],as_index=False).mean().rename(columns={'Price1':'MeanKwPrice1','Price2':'MeanKwPrice2'})
```

```python
df1 = pd.merge(all_df,keywords_pivot,on=['Keyword','file'])
```
