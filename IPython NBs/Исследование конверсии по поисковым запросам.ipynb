```python
parsedMetrikaData = []
for item in metrikaData['data']:
    searchPhrase = item['dimensions'][0]['name']
    visits = item['metrics'][0]
    bounces = item['metrics'][1]
    pageviews = item['metrics'][2]
    duration = visits*item['metrics'][3]
    parsedMetrikaData.append({'searchPhrase':searchPhrase,'visits':visits,'bounces':bounces,'pageviews':pageviews,'duration':duration})
```
