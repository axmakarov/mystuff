```python
def getMegaindexData(date,lr):
    url = 'http://api.megaindex.ru/?method=siteAnalyze&login=ax.makarov@yandex.ru&password=wae8Useisa&lr={1}&url=www.edemdom.ru&date={0}'.format(date,lr)
    data = requests.get(url).text
    df = pd.DataFrame(json.loads(data)['data'])
    df = df[[0,1]].drop(df.index[0]).reset_index(drop=True)
    df['date'] = date
    df['region'] = lr
    df.rename(columns={0:'query',1:'rank'})
    return df
```

```python
data = getMegaindexData("2016-04-03",213)
```

```python
allMegaindexDF = pd.DataFrame()
for date in dates:
    for lr in [1,213]:
        print '{0} {1}'.format(date,lr)
        try:
            data = getMegaindexData(date,lr)
            print 'Current:',len(data)
            allMegaindexDF = pd.concat([allMegaindexDF,data])
            print 'Total:',len(allMegaindexDF)
        except:
            print 'Error'
```
