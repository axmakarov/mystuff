```python
import json,requests
def getCalcData(minAdjustment,maxAdjustment):
    url = 'http://localhost:12345/AllInTimeTargetingService'
    headers = {'Content-Type': 'application/json'}
    payload = {
        "id":123, 
        "method": "OptimizeTimeTargeting", 
        "jsonrpc": "2.0", 
        "params": {
            "domain":"doorlock.ru",
            "startDate":"2015-10-01",
            "endDate":"2016-04-12",
            "minAdjustment":minAdjustment,
            "maxAdjustment":maxAdjustment
        }
    }
    return requests.post(url,json.dumps(payload),headers=headers).text
```

```python
def parseCalcData(data):
    json_data = json.loads(data)
    calcData = []
    for pair in json_data['result']['calcResult'].keys():
        pairData = json_data['result']['calcResult'][pair]
        pairData['weekdayHour'] = pair
        calcData.append(pairData)
    return calcData
```
