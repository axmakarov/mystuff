```python
from sklearn.cross_validation import train_test_split
X_train_cnv, X_test_cnv, y_train_cnv, y_test_cnv = train_test_split(array, cleared_df5['conversions'], test_size=0.5, random_state=32)
X_train_sum, X_test_sum, y_train_sum, y_test_sum = train_test_split(array, cleared_df5['sum'], test_size=0.5, random_state=32)
X_train_clk, X_test_clk, y_train_clk, y_test_clk = train_test_split(array, cleared_df5['clicks'], test_size=0.5, random_state=32)
```

Создаем регрессоры

```python
from sklearn import neighbors
knn_conversions = neighbors.KNeighborsRegressor(4)
knn_conversions.fit(X_train_cnv, y_train_cnv)
knn_sum = neighbors.KNeighborsRegressor(9)
knn_sum.fit(X_train_sum, y_train_sum)
knn_clicks = neighbors.KNeighborsRegressor(10)
knn_clicks.fit(X_train_clk, y_train_clk)

from sklearn import linear_model
br_conversions = linear_model.BayesianRidge()
br_conversions.fit(X_train_cnv, y_train_cnv)
br_sum = linear_model.BayesianRidge()
br_sum.fit(X_train_sum, y_train_sum)
br_clicks = linear_model.BayesianRidge()
br_clicks.fit(X_train_clk, y_train_clk)

from sklearn.ensemble import GradientBoostingRegressor
est_conversions = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=1, random_state=0, loss='ls')
est_conversions.fit(X_train_cnv, y_train_cnv)
est_sum = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=1, random_state=0, loss='ls')
est_sum.fit(X_train_sum, y_train_sum)
est_clicks = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=1, random_state=0, loss='ls')
est_clicks.fit(X_train_clk, y_train_clk)

from sklearn import tree
dtr_conversions = tree.DecisionTreeRegressor()
dtr_conversions.fit(X_train_cnv, y_train_cnv)
dtr_sum = tree.DecisionTreeRegressor()
dtr_sum.fit(X_train_sum, y_train_sum)
dtr_clicks = tree.DecisionTreeRegressor()
dtr_clicks.fit(X_train_clk, y_train_clk)

from sklearn import linear_model
sgd_conversions = linear_model.SGDRegressor()
sgd_conversions.fit(X_train_cnv, y_train_cnv)
sgd_sum = linear_model.SGDRegressor()
sgd_sum.fit(X_train_sum, y_train_sum)
sgd_clicks = linear_model.SGDRegressor()
sgd_clicks.fit(X_train_clk, y_train_clk)

from sklearn.svm import SVR
svr_conversions = SVR(C=1.0, epsilon=0.3)
svr_conversions.fit(X_train_cnv, y_train_cnv)
svr_sum = linear_model.SGDRegressor()
svr_sum.fit(X_train_sum, y_train_sum)
svr_clicks = linear_model.SGDRegressor()
svr_clicks.fit(X_train_clk, y_train_clk)
```

Рассчитываем точность

```python
print 'KNeighborsRegressor'
print 'Точность по конверсиям:', knn_conversions.score(X_test_cnv, y_test_cnv)
print 'Точность по расходу:', knn_sum.score(X_test_sum, y_test_sum)
print 'Точность по кликам:', knn_clicks.score(X_test_clk, y_test_clk)
print ' '
print 'BayesianRidge'
print 'Точность по конверсиям:', br_conversions.score(X_test_cnv, y_test_cnv)
print 'Точность по расходу:', br_sum.score(X_test_sum, y_test_sum)
print 'Точность по кликам:', br_clicks.score(X_test_clk, y_test_clk)
print ' '
print 'GradientBoostingRegressor'
print 'Точность по конверсиям:', est_conversions.score(X_test_cnv, y_test_cnv)
print 'Точность по расходу:', est_sum.score(X_test_sum, y_test_sum)
print 'Точность по кликам:', est_clicks.score(X_test_clk, y_test_clk)
print ' '
print 'DecisionTreeRegressor'
print 'Точность по конверсиям:', dtr_conversions.score(X_test_cnv, y_test_cnv)
print 'Точность по расходу:', dtr_sum.score(X_test_sum, y_test_sum)
print 'Точность по кликам:', dtr_clicks.score(X_test_clk, y_test_clk)
print ' '
print 'SGDRegressor'
print 'Точность по конверсиям:', sgd_conversions.score(X_test_cnv, y_test_cnv)
print 'Точность по расходу:', sgd_sum.score(X_test_sum, y_test_sum)
print 'Точность по кликам:', sgd_clicks.score(X_test_clk, y_test_clk)
print ' '
print 'SVR'
print 'Точность по конверсиям:', svr_conversions.score(X_test_cnv, y_test_cnv)
print 'Точность по расходу:', svr_sum.score(X_test_sum, y_test_sum)
print 'Точность по кликам:', svr_clicks.score(X_test_clk, y_test_clk)
```
