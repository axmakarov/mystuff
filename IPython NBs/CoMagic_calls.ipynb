```python
import json
import requests
import pandas as pd
```

```python
with open('D:\\calls.json','r+') as f1:
    comagicRawData = json.load(f1)
```

```python
df = pd.DataFrame(comagicRawData['data'])
df.to_csv('D:\\calltrack.csv',sep='\t',encoding='utf-8')
```

```python
df1 = df[df.duration > 0 & df['ac_id'].isin([24305,24811,17345,31882,30985,30726,22675,23750,31093])].groupby('site_id').count().reset_index()
df1.to_csv('D:\\t123.csv',sep='\t',encoding='utf-8')
```

```python
with open('D:\\callsSites.json','r+') as f1:
    comagicRawSitesData = json.load(f1)
```

```python
df2 = pd.DataFrame(comagicRawSitesData['data'])
df2 = df2.rename(columns={'id': 'site_id'})
df2
```

```python
pd.merge(df1, df2, on='site_id', how='outer').to_excel('D:\\callTrackingData.xlsx', sheet_name='Sheet1')
```
