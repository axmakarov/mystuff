```python
import sys
import json
consoleEncoding = 'cp1251'
questionsData = []
with open(u'C:\\Users\\aleksey.makarov\\Desktop\\2016-03\\Дерево стратегий\\Приложуха\\extra_questions.txt','r+') as f1:
    for row in f1:
        rowData = row.split('\t')
        question = rowData[0].decode('utf8')
        options = rowData[1].split(';')
        featureName = rowData[2].replace('\n','')
        options = [option.split(';') for option in options]
        optionDict = {}
        for option in options:
            option = option[0].replace(')','').replace('(','')
            optionData = option.split(',')
            optionNum = int(optionData[0])
            optionText = optionData[1].replace('"','').replace('\n','').decode('utf8')
            optionDict[optionNum] = optionText
        questionsData.append({'question':question,'options':optionDict,'featureName':featureName})
domainData = {}
skipLast = False
for item in questionsData:
    if item['question'] == questionsData[-1]['question'] and skipLast:
        continue
    while True:
        try:
            answer = int(raw_input((item['question'] + ' ').encode(consoleEncoding, errors='ignore')))
            domainData[item['featureName']] = item['options'][answer]
            break;
        except ValueError:
            sys.stdout.write(u'\nВведите число\n')
        except KeyError:
            sys.stdout.write(u'\nВведите число из списка доступных вариантов\n')
    if item['question'] == questionsData[2]['question'] and answer == 1:
        skipLast = True
        domainData[questionsData[-1]['featureName']] = '-'
with open('log\\{0}_{1}_domainData.json','w+') as f1:
    print >>f1, json.dumps(domainData).encode('utf8')
```
