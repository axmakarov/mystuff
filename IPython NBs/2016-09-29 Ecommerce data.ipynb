```python
import requests
import json
def GetToken(login,serviceType):
    if serviceType not in ['Metrika','Direct']:
        print 'Invalid service type'
        return None
    url = 'http://192.168.10.37:35678/TokenManagerService'
    payload = {'method':'GetToken','params':{'login':login.replace('@yandex.ru',''),'serviceType':serviceType}}
    r = requests.post(url,json.dumps(payload))
    data = json.loads(r.text)
    if len(data.keys()) == 1:
        print 'Invalid login'
        return None
    return data['result']['token']
```

```python
def getMetrikaCountersByToken(oAuthToken): #        
    url = 'https://beta.api-metrika.yandex.ru/management/v1/counters?oauth_token={0}'.format(oAuthToken)
    r = requests.get(url)
    try:
        response = json.loads(r.text)
        counters = response['counters']
        return counters
    except:
        return '-1'
```

```python
def getMetrikaReportInDataFrame(date1,date2,metrics,dimensions,token,counterId,filters,attribution):
    params = '&'.join(['dimensions={0}'.format(dimensions),
                       'metrics={0}'.format(metrics),
                       'date1={0}'.format(date1),
                      'date2={0}'.format(date2),
                      'ids={0}'.format(counterId),
                      'oauth_token={0}'.format(token),
                      'limit=100000',
                      'offset=1',
                      "filters={0}".format(filters),
                      'accuracy=full',
                      'attribution={0}'.format(attribution)])
    url = 'https://api-metrika.yandex.ru/stat/v1/data.csv?'+params
    print url
    df = pd.read_csv(url,encoding='utf-8')
    return df
```

```python
import psycopg2
def getVipProjectByDomain(domain):
    conn = psycopg2.connect("dbname='CRM' user='read_only' host='192.168.10.10' password='User_ro'")
    cur = conn.cursor()
    cur.execute("SELECT p.projectid, p.domain, p.status, p2.login, p2.password FROM project AS p INNER JOIN ProjectMetrikaData AS p2 ON p.UniqueId = p2.ProjectUniqueId WHERE domain=%s AND p.status<230 ORDER BY p.projectid", [domain])
    result = []
    rows = cur.fetchall()
    for row in rows:
        result.append(getVipProjectByRow(row))
    return result
def getVipProjectByRow(row):
    return {'projectid': row[0], 'domain': row[1], 'metrikalogin': row[3], 'metrikapassword': row[4]}
```

```python
domain = u'doorlock.ru'
def isThereAnyStatsByDomain(domain):
    project_info = getVipProjectByDomain(domain)
    if len(project_info) > 0:
        projectid = project_info[0]['projectid']
        visits = getSumUTMVisits(projectid)
        if visits > 0:
            return True,visits
        else:
            return False,0
    else:
        return False,0
```

```python
def getSumUTMVisits(projectid):
    conn = psycopg2.connect("dbname='CuboHistory' user='read_only' host='192.168.10.32' password='User_ro'")
    cur = conn.cursor()
    cur.execute("SELECT sum(utmvisits) FROM allinprojectstatistic WHERE projectid=%s", [projectid])
    rows = cur.fetchall()
    return rows[0][0]
```

```python
def getMetrikaEcomData(token,counterId):
    date1 = '30daysAgo'
    date2 = 'yesterday'
    metrics = 'ym:s:visits,ym:s:ecommerceRevenue,ym:s:ecommercePurchases,ym:s:productPurchasedQuantity'
    dimensions = ''
    filters = ''
    attribution = 'lastsign'
    metrika_df = getMetrikaReportInDataFrame(date1,date2,metrics,dimensions,token,counterId,filters,attribution)
    d30visits = int(metrika_df[u'""'])
    d30revenue = int(metrika_df[u''])
    d30purchases = int(metrika_df[u' '])
    d30quantity = int(metrika_df[u' '])
    date1 = '180daysAgo'
    date2 = 'yesterday'
    metrika_df = getMetrikaReportInDataFrame(date1,date2,metrics,dimensions,token,counterId,filters,attribution)
    d180visits = int(metrika_df[u'""'])
    d180revenue = int(metrika_df[u''])
    d180purchases = int(metrika_df[u' '])
    d180quantity = int(metrika_df[u' '])
    return d30visits,d30revenue,d30purchases,d30quantity,d180visits,d180revenue,d180purchases,d180quantity
```
