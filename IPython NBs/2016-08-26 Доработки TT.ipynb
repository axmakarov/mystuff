```python
import pandas as pd
def editDataFrame(df):
    calc_df = df
    calc_df.rename(columns={u'"День недели визита"': 'dayofweek',
                    u'Час визита':'hour',
                    u'Визиты': 'visits'
                   }, inplace=True)
    return calc_df
path = u'C:\\Users\\aleksey.makarov\\Desktop\\site.ru-Метки-UTM-2016-01-01-2016-08-17.csv'
df = pd.read_csv(path,encoding='utf-8-sig')
df = editDataFrame(df)
df.replace(to_replace={'dayofweek':{u'Воскресенье':u'Вс',
                                    u'Понедельник':u'Пн',
                                    u'Вторник':u'Вт',
                                   u'Среда':u'Ср',
                                   u'Четверг':u'Чт',
                                   u'Пятница':u'Пт',
                                   u'Суббота':u'Сб'}},inplace=True)
df = df.drop(df.index[0]).reset_index(drop=True)
```

```python
for i in xrange(0,len(df)):
    data = str(df.loc[i,'hour']).split(':')
    data[1] = str(int(data[0])+1)
    if len(data[1]) == 1:
        data[1] = '0'+data[1]
    df.loc[i,'hour'] = ' - '.join(data)
```

```python
hours = [u'00 - 01', u'01 - 02', u'02 - 03', u'03 - 04', u'04 - 05', u'05 - 06',
       u'06 - 07', u'07 - 08', u'08 - 09', u'09 - 10', u'10 - 11', u'11 - 12',
       u'12 - 13', u'13 - 14', u'14 - 15', u'15 - 16', u'16 - 17', u'17 - 18',
       u'18 - 19', u'19 - 20', u'20 - 21', u'21 - 22', u'22 - 23', u'23 - 24']
```

```python
days = [u'Пн', u'Вт', u'Ср', u'Чт', u'Пт', u'Сб', u'Вс']
```

```python
schedule = []
for hour in hours:
    for day in days:
        scheduleItem = {}
        scheduleItem['hour'] = hour
        scheduleItem['day'] = day
        scheduleItem['value'] = 100
        schedule.append(scheduleItem)
```

```python
schedule_df = pd.pivot_table(pd.DataFrame(schedule),values='value',index='day',columns='hour',aggfunc=numpy.sum,fill_value=100)
```
