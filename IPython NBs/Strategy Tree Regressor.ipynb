Группируем по каналу:

```python
df2 = df.groupby([u'Канал'],as_index=False).sum()
```

```python
df2
```

Удаляем такие каналы, у которых нулевой расход или переходы

```python
df2 = df2[(df2[u'Сумма по полю Расход, руб'] > 0) & (df2[u'Сумма по полю Переходы'] > 0)].reset_index(drop=True)
```

```python
df2
```

Формируем список допустимых каналов, выбросив "Итого" и ошибочные каналы

```python
channels = df2[~df2[u'Канал'].isin([u'Директ(Поиск)',u'Директ(РСЯ)',u'Итого'])][u'Канал'].reset_index(drop=True)
```

```python
channels
```

Формируем датафрейм с допустимыми каналами

```python
df3 = df[df[u'Канал'].isin(channels)]
```

Проверяем, что суммарные данные по каналам сходятся

```python
df3.groupby([u'Канал'],as_index=False).sum()
```

Удаляем строки данных без расхода или переходов

```python
df4 = df3[(df3[u'Сумма по полю Расход, руб'] > 0) & (df3[u'Сумма по полю Переходы'] > 0)]
```

```python
df4.groupby([u'Канал'],as_index=False).sum()
```

Формируем датафрейм для обучения с нужными колонками

```python
df5 = df4[[u'Канал',u'Сумма по полю Расход, руб',u'Сумма по полю Конверсии',u'Сумма по полю Переходы',u'Сумма по полю Кол-во звонков',u'Витрина товаров',u'Опт/розница',u'Товары / услуги',u'Указаны ли на сайте цены',u'Сформированность спроса',u'B2B/B2C']].reset_index(drop=True)
```

Переименовываем колонки

```python
df5.rename(columns={u'Канал': 'channel',
                    u'B2B/B2C': 'b2b',
                    u'Сумма по полю Расход, руб': 'sum', 
                    u'Сумма по полю Переходы': 'clicks', 
                    u'Сумма по полю Конверсии': 'conversions', 
                    u'Сумма по полю Кол-во звонков': 'calls',
                    u'Витрина товаров': 'showcase',
                    u'Опт/розница': 'wholesale',
                    u'Товары / услуги': 'goods/services',
                    u'Указаны ли на сайте цены': 'isPrices',
                    u'Сформированность спроса': 'demand'
                   }, inplace=True)
```

```python
df5.head(15)
```

Формируем датафрейм с обучающими данными

```python
train_df = df5[['channel','showcase','wholesale','goods/services','isPrices','demand','b2b']]
```

Из обучающего датафрейма делаем список словарей

```python
list_of_dictionaries = [train_df.iloc[line,:].T.to_dict() for line in range(len(train_df))]
```

Делаем массив фич

```python
from sklearn.feature_extraction import DictVectorizer
vec = DictVectorizer()
```

```python
array = vec.fit_transform(list_of_dictionaries).toarray()
```

Получаем названия фич

```python
vec.get_feature_names()
```

Создаем регрессоры

```python
from sklearn import neighbors
knn_conversions = neighbors.KNeighborsRegressor(4)
knn_conversions.fit(array, df5['conversions'])
knn_sum = neighbors.KNeighborsRegressor(9)
knn_sum.fit(array, df5['sum'])
knn_clicks = neighbors.KNeighborsRegressor(10)
knn_clicks.fit(array, df5['clicks'])

from sklearn import linear_model
br_conversions = linear_model.BayesianRidge()
br_conversions.fit(array, df5['conversions'])
br_sum = linear_model.BayesianRidge()
br_sum.fit(array, df5['sum'])
br_clicks = linear_model.BayesianRidge()
br_clicks.fit(array, df5['clicks'])

from sklearn.ensemble import GradientBoostingRegressor
est_conversions = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=1, random_state=0, loss='ls')
est_conversions.fit(array, df5['conversions'])
est_sum = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=1, random_state=0, loss='ls')
est_sum.fit(array, df5['sum'])
est_clicks = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, max_depth=1, random_state=0, loss='ls')
est_clicks.fit(array, df5['clicks'])

from sklearn import tree
dtr_conversions = tree.DecisionTreeRegressor()
dtr_conversions.fit(array, df5['conversions'])
dtr_sum = tree.DecisionTreeRegressor()
dtr_sum.fit(array, df5['sum'])
dtr_clicks = tree.DecisionTreeRegressor()
dtr_clicks.fit(array, df5['clicks'])

from sklearn import linear_model
sgd_conversions = linear_model.SGDRegressor()
sgd_conversions.fit(array, df5['conversions'])
sgd_sum = linear_model.SGDRegressor()
sgd_sum.fit(array, df5['sum'])
sgd_clicks = linear_model.SGDRegressor()
sgd_clicks.fit(array, df5['clicks'])

from sklearn.svm import SVR
svr_conversions = SVR(C=1.0, epsilon=0.3)
svr_conversions.fit(array, df5['conversions'])
svr_sum = linear_model.SGDRegressor()
svr_sum.fit(array, df5['sum'])
svr_clicks = linear_model.SGDRegressor()
svr_clicks.fit(array, df5['clicks'])
```

Рассчитываем точность

```python
print 'KNeighborsRegressor'
print 'Точность по конверсиям:', knn_conversions.score(array, df5['conversions'])
print 'Точность по расходу:', knn_sum.score(array, df5['sum'])
print 'Точность по кликам:', knn_clicks.score(array, df5['clicks'])
print ' '
print 'BayesianRidge'
print 'Точность по конверсиям:', br_conversions.score(array, df5['conversions'])
print 'Точность по расходу:', br_sum.score(array, df5['sum'])
print 'Точность по кликам:', br_clicks.score(array, df5['clicks'])
print ' '
print 'GradientBoostingRegressor'
print 'Точность по конверсиям:', est_conversions.score(array, df5['conversions'])
print 'Точность по расходу:', est_sum.score(array, df5['sum'])
print 'Точность по кликам:', est_clicks.score(array, df5['clicks'])
print ' '
print 'DecisionTreeRegressor'
print 'Точность по конверсиям:', dtr_conversions.score(array, df5['conversions'])
print 'Точность по расходу:', dtr_sum.score(array, df5['sum'])
print 'Точность по кликам:', dtr_clicks.score(array, df5['clicks'])
print ' '
print 'SGDRegressor'
print 'Точность по конверсиям:', sgd_conversions.score(array, df5['conversions'])
print 'Точность по расходу:', sgd_sum.score(array, df5['sum'])
print 'Точность по кликам:', sgd_clicks.score(array, df5['clicks'])
print ' '
print 'SVR'
print 'Точность по конверсиям:', svr_conversions.score(array, df5['conversions'])
print 'Точность по расходу:', svr_sum.score(array, df5['sum'])
print 'Точность по кликам:', svr_clicks.score(array, df5['clicks'])
```

Прогоняем на тестовых данных

```python
testData = {'channel': u'Директ',
           'b2b': u'B2C',
           'demand': u'Да',
           'goods/services': u'Услуги',
           'isPrices': u'Да',
           'showcase': u'Да',
           'wholesale': u'-'}
point = vec.transform(testData).toarray()

print 'KNeighborsRegressor'
print 'Conversion:', knn_conversions.predict(point)[0]
print 'Sum:', knn_sum.predict(point)[0]
print 'Clicks:',knn_clicks.predict(point)[0]
print 'CPC:',knn_sum.predict(point)[0]/knn_clicks.predict(point)[0]
print 'Conversion:',knn_conversions.predict(point)[0]/knn_clicks.predict(point)[0]*100

print ' '
print 'DecisionTreeRegressor'
print 'Conversion:', dtr_conversions.predict(point)[0]
print 'Sum:', dtr_sum.predict(point)[0]
print 'Clicks:',dtr_clicks.predict(point)[0]
print 'CPC:',dtr_sum.predict(point)[0]/dtr_clicks.predict(point)[0]
print 'Conversion:',dtr_conversions.predict(point)[0]/dtr_clicks.predict(point)[0]*100

print ' '
print 'Сonfidence:',dtr_conversions.predict(point)[0]/dtr_clicks.predict(point)[0]*100-knn_conversions.predict(point)[0]/knn_clicks.predict(point)[0]*100

c1 = dtr_conversions.predict(point)[0]/dtr_clicks.predict(point)[0]*100
c2 = knn_conversions.predict(point)[0]/knn_clicks.predict(point)[0]*100

if abs(c1-c2) < 0.5:
    c3 = (c1+c2)/2
elif ((c1 == 0.0) | (c2 == 0.0)):
    c3 = 0
else:
    c3 = min(c1,c2)
    
print 'Total conversion:',c3
```

```python
allStats = []
for channel in df5['channel'].unique().tolist():
    for b2b in df5['b2b'].unique().tolist():
        for goodsServices in df5['goods/services'].unique().tolist():
            for wholesale in df5['wholesale'].unique().tolist():
                for showcase in df5['showcase'].unique().tolist():
                    for isPrices in df5['isPrices'].unique().tolist():
                        for demand in df5['demand'].unique().tolist():
                            testData = {'channel': channel,
           'b2b': b2b,
           'demand': demand,
           'goods/services': goodsServices,
           'isPrices': isPrices,
           'showcase': showcase,
           'wholesale': wholesale}
                            point = vec.transform(testData).toarray()
                            testData['knn_conversions'] = knn_conversions.predict(point)[0]
                            testData['knn_sum'] = knn_sum.predict(point)[0]
                            testData['knn_clicks'] = knn_clicks.predict(point)[0]
                            testData['knn_cpc'] = knn_sum.predict(point)[0]/knn_clicks.predict(point)[0]
                            testData['knn_cr'] = knn_conversions.predict(point)[0]/knn_clicks.predict(point)[0]*100
                            testData['dtr_conversions'] = dtr_conversions.predict(point)[0]
                            testData['dtr_sum'] = dtr_sum.predict(point)[0]
                            testData['dtr_clicks'] = dtr_clicks.predict(point)[0]
                            testData['dtr_cpc'] = dtr_sum.predict(point)[0]/dtr_clicks.predict(point)[0]
                            testData['dtr_cr'] = dtr_conversions.predict(point)[0]/dtr_clicks.predict(point)[0]*100
                            testData['confidence'] = abs(testData['dtr_cr'] - testData['knn_cr'])
                            testData['datapoints'] = df5[(df5['channel'] == testData['channel']) &
   (df5['b2b'] == testData['b2b']) &
   (df5['demand'] == testData['demand']) &
   (df5['goods/services'] == testData['goods/services']) &
   (df5['isPrices'] == testData['isPrices']) &
   (df5['showcase'] == testData['showcase']) &
   (df5['wholesale'] == testData['wholesale'])]['channel'].count()
                            allStats.append(testData)
```
