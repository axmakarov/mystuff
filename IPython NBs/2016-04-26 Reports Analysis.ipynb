```python
import pandas as pd
from os import listdir
from os.path import isfile, join
import pandas_profiling
```

Идентификатор критерия 3000006 указывает на статистику, связанную с Оптимизатором кампаний в КМС.

Идентификатор критерия 3000000 указывает на статистику, связанную с показами в контекстно-медийной сети

https://developers.google.com/adwords/api/docs/guides/reporting-concepts#multiple

```python
mypath = u'C:\\Users\\aleksey.makarov\\Desktop\\2016-04\\AdWords\\AdwordsReports\\'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
keywords_stats = pd.DataFrame()
slots_stats = pd.DataFrame()
for f in onlyfiles:
    if '1.csv' in f:
        domain = f.replace(u'-1.csv','')
        df = pd.read_csv(mypath+f,header=1,skipfooter=1)
        df['domain'] = domain
        keywords_stats = pd.concat([keywords_stats,df])
    if '2.csv' in f:
        domain = f.replace(u'-2.csv','')
        df = pd.read_csv(mypath+f,header=1,skipfooter=1)
        df['domain'] = domain
        slots_stats = pd.concat([slots_stats,df])
keywords_stats = keywords_stats.reset_index(drop=True)
slots_stats = slots_stats.reset_index(drop=True)
keywords_stats_without_display_network = keywords_stats[~keywords_stats['Keyword ID'].isin([3000006,3000000])].reset_index(drop=True)
```

```python
keywords_stats.to_excel('C:\\Users\\aleksey.makarov\\Desktop\\2016-05\\AdWordsSNDN.xlsx')
```

```python
pandas_profiling.ProfileReport(keywords_stats_without_display_network)
```

```python
keywords_stats_without_display_network = keywords_stats_without_display_network[keywords_stats_without_display_network['Max. CPC'] != ' --'].reset_index(drop=True)
keywords_stats_without_display_network['Max. CPC'] = keywords_stats_without_display_network['Max. CPC'].str.replace('auto: ','')
```

```python
keywords_stats_without_display_network['Max. CPC'] = keywords_stats_without_display_network['Max. CPC'].fillna(0)
keywords_stats_without_display_network['Max. CPC'] = keywords_stats_without_display_network['Max. CPC'].astype(int)
```

```python
for i in xrange(0,len(keywords_stats_without_display_network)):
    try:
        if keywords_stats_without_display_network.loc[i,'Max. CPC'] == 0:
            keywords_stats_without_display_network.loc[i,'Max. CPC'] = keywords_stats_without_display_network.loc[i,'Avg. CPC']
    except:
        pass
```

```python
keywords_stats_without_display_network.to_excel(u'C:\\Users\\aleksey.makarov\\Desktop\\2016-04\\AdWords\\adwords_stats.xlsx')
```

```python
slots_stats_for_merge = slots_stats[slots_stats['Top vs. Other'] == 'Google search: Top'][['Ad group ID','Campaign ID','Keyword ID','Cost','Clicks','Impressions']].rename(columns={
        'Cost':'Top Slot Cost',
        'Clicks':'Top Slot Clicks',
        'Impressions':'Top Slot Impressions'
    })
```

```python
merged_adwords_stats = pd.merge(keywords_stats_without_display_network,slots_stats_for_merge,how='left',on=['Ad group ID','Campaign ID','Keyword ID'])
```

```python
merged_adwords_stats.to_excel(u'C:\\Users\\aleksey.makarov\\Desktop\\2016-04\\AdWords\\adwords_stats.xlsx')
```
