```python
import pandas as pd
from os import listdir
from os.path import isfile, join
mypath = u'\\\CHERNIKOVA-IDEV\\Share\\AdPrice\\'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
all_df = pd.DataFrame()
for f in onlyfiles:
    if '.csv' in f:
        df = pd.read_csv(mypath+f,sep=';',encoding='cp1251')
        df['file'] = f
        all_df = pd.concat([df,all_df])
```

```python
all_df
```

```python
OurCampaignsDataFrame = all_df[~all_df['Url'].isin(['2k17.ru'])]
```

```python
PetuhDataFrame = all_df[all_df['Url'] == '2k17.ru']
```

```python
PetuhDataFrame.rename(columns={'Price1':'2k17_Price1','Price2':'2k17_Price2'},inplace=True)
```

```python
PetuhDataFrame = PetuhDataFrame[['CampaignExternalId','AdExternalId','KeyPhraseExternalId','2k17_Price1','2k17_Price2']]
```

```python
super_df = pd.merge(OurCampaignsDataFrame,PetuhDataFrame,on=['CampaignExternalId','AdExternalId','KeyPhraseExternalId'],how='left').dropna()
```

```python
super_df.to_excel('D:\\karma_our_ads.xlsx')
```
